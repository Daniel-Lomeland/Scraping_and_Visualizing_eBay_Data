# <center> Collecting eBay Listings Data

<center> This script creates a CSV file of ebay listings information.

<center> Output columns in the CSV file:

<center>- Title
<center>- Price
<center>- End date
<center>- End time
<center>- Shipping costs
<center>- A link to item page


## <center> Important 

<center> This script is made to search sold ebay listings.
    
<center> Make sure you  are looking at completed listings and that you do
    
<center>all your search filtering on ebay befor you collect your input URL.

```python
# How many results do you want?

desired_number_of_results = 54


# What is the URL of the first page of listings you want to scrape?

input_link = 'https://www.ebay.com/sch/39472/i.html?_from=R40&_nkw=1922&LH_Complete=1&rt=nc&LH_BIN=1'


# Where Would you like to Save your CSV File?
# (If path is inputed the file will be saved in the same location as this notebook)

csv_path_and_title = 'ebay_1922gold'
```

 <center> Don't forget to run input cell

# <center> Time To Start Scraping
 <center> Now run cell below to get results!
    

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import math

# Prepare inputs

partial_link = input_link[:-1]

d = desired_number_of_results/50
ru = math.ceil(d)

csv_path_and_title = f'{csv_path_and_title}.csv'

# Loop through ebay listings.

n = 0
ebay_listings = []

for r in range(ru):
    
    n = n+1
    url = f'{partial_link}{str(n)}' 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    results = soup.find_all('li', class_='s-item')
    
    for result in results:    
                title = result.find('h3', class_="s-item__title")

                title1 = list(title.strings)[-1]
                status = list(title.strings)[0]

                s_split= status.split(" ")
                sold= s_split[0]


                price = result.find('div', class_="s-item__details").span.text
                end_time = result.find('span', class_="s-item__ended-date").text
                time = end_time[7:]
                date = end_time[:6]
                shipping = result.find('span', class_="s-item__shipping").text
                link = result.a['href']
                
                #Add to list of listings
                ebay_listings.append({'title': title1,
                                      'result': status,
                                      'price': price,
                                      'sold': sold,
                                      'end_date':date,
                                      'end_time':time,
                                      'shipping':shipping,
                                      'link':link              
                                })

                
df = pd.DataFrame(ebay_listings)

df = df.iloc[:desired_number_of_results]

ebay_listings_df = df[['title','price','sold','result','end_date','end_time','link']]

ebay_listings_df.to_csv(csv_path_and_title)

ebay_listings_df
```
